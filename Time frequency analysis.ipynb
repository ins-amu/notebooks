# Time frequency analysis

This notebook shows how to do simple time-frequency analysis using
basic Python scientific libraries.

This notebook plots figures in-line, so we first need to enable this:

```python
%pylab inline
```

We'll use the following common libraries

```python
import numpy as np
import pylab as pl
from scipy import signal
```

## Synthesizing a simple signal

Create a signal containing both noise and nonstationary frequency content:

```python
# 30 seconds at 1024 Hz
fs = 1024.0
dt = 1.0 / fs
t = np.r_[:30.0:dt]

#   sin starts at 15.0 Hz, increases in time + Gaussian noise
y = np.sin(t * 2 * np.pi * (5.0 + t)) + np.random.randn(*t.shape) * 2.0
```

Now, plot the first and last second of the signal:

```python
subplot(211)
pl.plot(t[:1024], y[:1024], 'k')
subplot(212)
pl.plot(t[-1024:], y[-1024:], 'k')
xlabel('Time (s)');
```

## Analysis

A basic form of time-frequency analysis will

- split the signal into windows
- apply a window function
- fft each window

```python
# size of FFT
nfft = 256

# make window function
win = signal.hanning(nfft)
pl.plot(t[:win.size], win)
pl.ylabel('Hanning window')
pl.xlabel('Time (s)');
```

```python
# windowed y
yw = y.reshape((-1, nfft)) * win
# FFT 
Fy = np.fft.fft(yw, axis=-1)
# compute frequency bins
f = np.fft.fftfreq(nfft, dt)
```

After computing the frequency domain representation, we can choose a frequency band and plot the image:

```python
# choose freq band to look at
mask = (f > 0.0) * (f < 100.0)
Fym = Fy[:, mask]
fm = f[mask]

pl.imshow(np.abs(Fym).T[::-1], extent=[t[0], t[-1], fm[0], fm[-1]], aspect='auto')
```
